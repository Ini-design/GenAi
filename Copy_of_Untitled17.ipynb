```python
import gradio as gr
import torch
from transformers import pipeline
```

```python
text_generate = pipeline("text-generation",
 model="sshleifer/distilbart-cnn-12-6",
  torch_dtype = torch.bfloat16)
```

```python
def generate(input):

    output= text_generate(input)
    return output [0] ['text_generate']
```

```python
gr.close_all()


demo = gr.Interface(
fn=generate,
inputs =[ gr.Textbox(label="input the text and get more and the topics",lines=5)],
outputs = [gr.Textbox(label="the generated text", lines=3)],
description="Ai assistants "
)
demo.launch(debug= True)
```
